# nedir şu patch olayı:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
  
# The image
x = np.arange(16).reshape(4, 4)
  
# highlight some feature in the
# middle boxes.
fig, ax = plt.subplots()

ax.imshow(x, cmap = plt.cm.gray,
          interpolation ='nearest');
```

```python
# şimdi buna patch ekleyelim: 

x = np.arange(16).reshape(4, 4)

fig, ax = plt.subplots()

ax.imshow(x, cmap = plt.cm.gray,
          interpolation ='nearest')

ax.add_patch( Rectangle((0.5, 0.5),
                        2, 2,
                        fc ='none', 
                        ec ='g',
                        lw = 10) );
```

```python
fig, ax = plt.subplots(dpi = 100)

sns.countplot(x='season', data=df);

for p in ax.patches:
    ax.annotate("text", (p.get_x()+0.3, p.get_height()+20));
    
# ax.annotate -> ilk parametrede yazdırılacak texti belirliyoruz.
# 2. parametre ile bu text'in x ve y olarak hangi konuma yazdırılacağı belirliyoruz.
```
